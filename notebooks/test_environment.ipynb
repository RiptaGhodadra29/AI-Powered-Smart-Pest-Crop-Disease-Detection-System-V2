```python
import matplotlib.pyplot as plt

images, labels = next(iter(train_loader))

print("Images Shape:", images.shape)
print("Labels Shape:", labels.shape)

fig, axes = plt.subplots(2, 4, figsize=(12, 6))

for i, ax in enumerate(axes.flat):

    image = images[i].permute(1, 2, 0).numpy()

    image = (
        image - image.min()
    ) / (
        image.max() - image.min()
    )

    ax.imshow(image)

    ax.set_title(
        f"Label: {labels[i].item()}"
    )

    ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn as nn

from src.models.efficientnet_b0 import EfficientNetB0
from src.training.trainer import Trainer
from src.training.utils import EarlyStopping
from torch.optim import AdamW
from torch.optim.lr_scheduler import CosineAnnealingLR

# Device
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")

# Model
model = EfficientNetB0(num_classes=15).to(device)

# Loss
criterion = nn.CrossEntropyLoss()

# Optimizer
optimizer = AdamW(model.parameters(), lr=1e-3, weight_decay=1e-4)

# Scheduler
scheduler = CosineAnnealingLR(optimizer, T_max=10)

# Use the real dataset loaders defined earlier in the notebook
# (make sure create_dataloaders has already been executed)
# train_loader, val_loader, test_loader = create_dataloaders(...)

# Early stopping
early_stopping = EarlyStopping(patience=5)

# Trainer
trainer = Trainer(
    model,
    train_loader,
    val_loader,
    optimizer,
    criterion,
    scheduler,
    device
)

# Train
trainer.fit(epochs=20, early_stopping=early_stopping)
```
